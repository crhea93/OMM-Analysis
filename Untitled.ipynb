```python
import matplotlib.pyplot as plt

from astropy.wcs import WCS
from astropy.io import fits
from astropy.utils.data import get_pkg_data_filename
import numpy as np
from astropy import units as u
from astropy.coordinates import Angle
```

```python
filename = '/home/carterrhea/Dropbox/OMM/200611/Target/CIZAJ1824.1+3029/200611_0000001297.fits'

hdu = fits.open(filename)[0]
coords = WCS(hdu.header)
print(coords.wcs.naxis)
plt.subplot(projection=coords)
plt.imshow(np.log10(hdu.data), origin='lower')
#plt.grid(color='white', ls='solid')
plt.xlabel('RA')
plt.ylabel('Dec')
```

```python
hdu.header
```

```python
pixel = Angle(0.459211, u.arcsec)
pixel.degree
```

```python
u.set_enabled_equivalencies(u.dimensionless_angles())
u.deg.to('') 

coords.wcs.crpix = [512, 272] # center pixel

coords.wcs.crval = [hdu.header['RA'], hdu.header['DEC']] # RA and dec values in hours and degrees

coords.wcs.ctype = ["", ""]

coords.wcs.cdelt = [pixel.degree, pixel.degree]
coords
```

```python
plt.subplot(projection=coords)
plt.imshow(np.log10(hdu.data), origin='lower')
#plt.grid(color='white', ls='solid')
plt.xlabel('RA')
plt.ylabel('Dec')
```

```python
header = coords.to_header()
hdu = fits.PrimaryHDU(data=hdu.data, header=header)
hdu.writeto('test.fits', overwrite=True)
```
